```python
# Define N (number of sectors)
N = 30
```

```python
# def corresponding sector list for comp_name list
sector_list = list(name_sector['sector'])
```

```python
#def condition functions on m

#Returns a bool that check if matched_indices condition is satisfied.
# comp_number_list is the list for a specific company in s a single time period. The dimension of the list is mx1
# matched_indices is the indices list we want to match
# For example, matched_indices can be [1,3,4], which denote that for this company, it must satisfy certain properties for corresponding datasets
# bounded_value is the cut-off constant c that we want comp_number_list[i-1] > c for all i in matched_indices. Note we take i-1 because list starts with list[0]
def matched_indices_condition(comp_number_list, bounded_value=0, matched_indices=[]):
    b = True
    for i in matched_indices:
        s = i-1
#         print(comp_number_list)
        new = comp_number_list[s] > bounded_value
        b = b and new
    return b

#Again returns a bool. Now we want at least total_number in comp_number_list to be greater than bounded value.
def total_number_condition(comp_number_list, bounded_value=0, total_number=1):
    count=0
    for comp_number in comp_number_list:
        if comp_number>bounded_value:
            count+=1
    return count>= bounded_value

# functions work on a single time period t. E.g, t=19800101. 
# Returns 1xN array D where D[i] counts for the corresponding satisfing number for sector i
def single_period_func(time,bounded_value=0, matched_indices=[],total_number=1):
    array = np.zeros(N)
    df = pd.read_csv("output"+str(time)+".csv")
    for comp in name_list:
        comp_dataframe = df[str(comp)]
#         print(comp_dataframe)
        comp_number_list = list(comp_dataframe)
        
        trueA = matched_indices_condition(comp_number_list, bounded_value, matched_indices)
        trueB = total_number_condition(comp_number_list, bounded_value, total_number)
        if trueA and trueB:
            index = name_list.index(comp)
            sector = sector_list[index]
            array[sector-1] += sum(comp_number_list)
    return array

#Returns TxN matrix A for all time periods where A[t][n] is the the number for sector n in time t
def all_period_func(bounded_value=0, matched_indices=[], total_number=1):
    A = []
    for time in timeseries:
        array = single_period_func(time,bounded_value, matched_indices,total_number)
        A.append(array)
        print(time)
    return A
```

```python
A =  all_period_func()
arrayA = np.array(A)
df = pd.DataFrame(arrayA)
print(df)

sec_list=[]
for i in range(N):
    sec_list.append('sector'+str(i+1))
df.columns = sec_list
df.to_csv("output_matrix_A.csv")
```
